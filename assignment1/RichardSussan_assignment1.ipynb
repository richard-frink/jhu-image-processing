```python
from skimage import data, io, filters
from skimage.io import imsave
from skimage.draw import polygon
import numpy as np
```

```python
image = io.imread('colorwheel.png')

# using filled polygon to create the E
poly = np.array((
    (100, 100),
    (100, 230),
    (125, 230),
    (125, 125),
    (175, 125),
    (175, 200),
    (200, 200),
    (200, 125),
    (250, 125),
    (250, 230),
    (275, 230),
    (275, 100),
    (100, 100),
))
rr, cc = polygon(poly[:, 0], poly[:, 1], image.shape)
image[rr, cc, :] = [0,255,255]

# show the image
io.imshow(image)
imsave("colorwheel_with_e.jpeg", image)
```
